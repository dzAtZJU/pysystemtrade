```python
from paper.systems.simplesystem import simplesystem
from paper.topic.common_import import *

vol_look_back = 35 * 24
look_back = np.timedelta64(15 * 365 + vol_look_back//24 , "D")
end = pd.Timestamp.now()
start = end - look_back
period = (start, end)
fre = 'Hour'

system = simplesystem(config='paper.systems.global.yaml')
cnh = 1 / select_period(system.data.get_raw_price('CNH').rename('CNH'), period)
cnh_onshore = select_period(system.data.get_raw_price('CNH-onshore').rename('CNH-onshore'), period)
JPY_mini = select_period(system.data.get_raw_price('JPY_mini').rename('JPY_micro'), period)
EUR_micro = select_period(system.data.get_raw_price('EUR_micro').rename('EUR_micro'), period)
GBP_micro = select_period(system.data.get_raw_price('GBP_micro').rename('GBP_micro'), period)
BTC = select_period(artic_lib.read('{}/{}'.format(fre, 'BTC-USDT-Binance')).data['FINAL'].rename('BTC'), period)
US2 = select_period(system.data.get_raw_price('US2').rename('US2'), period)
GOLD = select_period(system.data.get_raw_price('GOLD').rename('GOLD'), period)
```

```python
_ = cnh.plot(figsize=(16,5), legend=True)
_ = cnh_onshore.plot(legend=True)
plt.show()
_ = JPY_mini.plot(figsize=(16,5), legend=True)
plt.show()
_ = EUR_micro.plot(figsize=(16,5), legend=True)
plt.show()
_ = GBP_micro.plot(figsize=(16,5), legend=True)
plt.show()
_ = BTC.plot(figsize=(16,5), legend=True)
plt.show()
_ = US2.plot(figsize=(16,5), legend=True)
plt.show()
_ = GOLD.plot(figsize=(16,5), legend=True)
plt.show()
```
