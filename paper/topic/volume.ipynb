```python
period = ('2020-01-01', pd.Timestamp.now())
ins = 'ETH-USDT-Binance'

# Data
history = artic_lib.read('Hour/{}'.format(ins)).data
volume = select_period(history['VOLUME'], period)
close = select_period(history['FINAL'].rename('close price'), period)

# Feature & Views
feature1 = (close/close.iloc[0])
feature1.plot(legend=True, figsize=(16, 4), style='.', ms=1)
# volume.plot(secondary_y=True, legend=True, style='.')
plt.title('{} {}'.format(ins, period))
plt.show()
show(select_period(feature1, period).describe(), caption = 'Distribution')
print(2.16 / 9.5 - 1)
show(feature1)
```
