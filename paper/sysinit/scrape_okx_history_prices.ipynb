```python
from arctic import Arctic
from itables import show
import matplotlib.pyplot as plt

store = Arctic('localhost')
library = store['production.perpetual_prices_v3']
p1 = library.read('Day/LUV').data['FINAL'].rename('LUV').tail(500)#.pct_change().dropna()
p1.plot(figsize=(15,6), legend=True)

p2 = library.read('Day/AAL').data['FINAL'].rename('AAL').tail(500)#.pct_change().dropna()
p2.plot(legend=True)

p3 = library.read('Day/UAL').data['FINAL'].rename('UAL').tail(500)#.pct_change().dropna()
p3.plot(legend=True)

import pandas as pd
display(pd.concat([p1, p2, p3], axis=1).pct_change().corr())
```
